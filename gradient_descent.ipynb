# Gradient Descent


```python
from sklearn.datasets import make_regression
```

```python
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_    #m
```

```python
reg.intercept_  #b
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# now  let us apply Gradient Descent assuming slope is constant m=78.35
# let us assume initial value of intercept b=0;
```

```python
y_pred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
```

```python
m=78.35
b=0

loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
#learning rate=0.1
lr=0.1
step_size=loss_slope*0.1
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color="black",label='b={}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
```

```python
#iteration 2
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python

step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color="black",label='b={}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
```
