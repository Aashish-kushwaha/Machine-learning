```python
# function to return the team particular batsman has scored most runs
def scored_runs(batsman_name):
    mask1=df['batsman']==batsman_name
    bt=df[mask1]
    return bt.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
scored_runs('V Kohli')
```

## isin() function

```python
# find the most destructive batsman in death over in th history of IPL
#strike rate 
#min batsman 200 balls in over 16-20
```

```python
mask=df['over']>15  #death over
death_overs=df[mask]
```

```python
#find number of balls played by every batsman
allbt=death_overs.groupby('batsman')['batsman_runs'].count() #count of ball played by every batsman
mask=allbt>200          #batsman who played more than 200 balls
batsman_list=allbt[mask].index.tolist()   #list of batsman
len(batsman_list)
```

```python
#to find strike rate
#1.runs scored by all these 43 batsman
#2.balls played by batsman
final=df[df['batsman'].isin(batsman_list)]  #we check for every batsman is present in this list
```

```python
#now in this final dataframe we need to calculate 
#1.runs scored
#2.balls played by these 43 batsman
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

## merge function 
it join two data frame: hte condition is both dataframe should have something in common

```python
df=pd.read_csv('deliveries.csv')
match=pd.read_csv('data.csv')
df.shape
```

```python
#list of orange cap holder:the batsman who hit most run in a single season
#id is common in both data.csv and deliveries.csv
new=df.merge(match,left_on='match_id',right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```
