```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])

```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("rohit sharma vs virat kohli")
plt.xlabel('season of ipl')
plt.ylabel('runs scored')
```

```python
#change colors
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linewidth=2)

plt.title("rohit sharma vs virat kohli")
plt.xlabel('season of ipl')
plt.ylabel('runs scored')

```

```python
#see marker
#change colors
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linewidth=2,marker='o')

plt.title("rohit sharma vs virat kohli")
plt.xlabel('season of ipl')
plt.ylabel('runs scored')

```

```python
#name line
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dotted',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linewidth=2,label='rohit')

plt.title("rohit sharma vs virat kohli")
plt.xlabel('season of ipl')
plt.ylabel('runs scored')

plt.legend(loc='best')
```

```python
# listing axis
price=[50000,34000,23444,90000,28333222] #outlier
year=[2015,2016,2018,2019,2020]

plt.plot(year,price)
# to remove flatness due to outlier
plt.ylim(0,100000)  #we trim the outlier
```

## bar chart
1.bivariate analysis\
2.numerical vs categorical\
3.use case- Aggregate analysis of groups

```python
#simple bar chart
children=[10,20,30,40,50]
color=['red','green','blue','black','pink']
```

```python
plt.bar(color,children,color='black')
plt.xlabel('color')
plt.ylabel('children')
```

```python
#horizontal barchart
plt.barh(color,children,color='black')
plt.xlabel('color')
plt.ylabel('children')
```

```python
#multiple barchart
df=pd.read_csv('batsman_season_record.csv')

df.head()
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width=0.3)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.3)
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width=0.3)


#to show name of batsman
plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
#stacked Bar chart
plt.bar(df['batsman'],df['2015'],color='red')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'])
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'])
```
