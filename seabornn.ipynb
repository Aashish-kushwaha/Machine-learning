## Scatter Plot

axes level

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express  as px
```

```python
tips=pd.read_csv('tips.csv')
```

```python
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='smoker',style='sex',size='size')
```

##  make scatter as figure level


 using replot

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='smoker',style='sex',size='size')
```

## it is better to use figure level as we make multiple axes level 

## line plot

```python
gap=px.data.gapminder()
```

```python
gap
```

```python
#filter data on country code
mask=gap['country']=='India'
```

```python
temp_df=gap[mask]
```

```python
temp_df
```

```python
#axes level
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
#using Relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
sns.lineplot(data=gap,x='year',y='lifeExp',hue='country')
```

```python
##filter data for india pakistan and china
mask=gap['country'].isin(['India','Pakistan','China'])
```

```python
df=gap[mask]
```

```python
df
```

```python
sns.relplot(data=df,x='year',y='lifeExp',kind='line',hue='country')
```

## Histogram

```python
# figure level -> displot
# axes level histplot
```

```python
sns.histplot(data=tips,x='total_bill',bins=10)
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',hue='sex')
```

```python
sns.displot(data=tips,x='day',kind='hist')
```

```python
titanic=pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
#facet col and row
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex',col='sex')

```

## KDE plot

```python
#uses gaussian kernel to smoothen the function

sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='tip',kind='kde',hue='sex')
```

## Rugplot

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```


## Bivariate histogram

```python
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
## bivariate kde
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

## heatmap

```python
df=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
sns.heatmap(df)
```

```python
d=gap[gap['continent']=='Europe']
```

```python
df=d.pivot(index='country',columns='year',values='lifeExp')
```

```python
sns.heatmap(df,annot=True,linewidths=0.5)
```
